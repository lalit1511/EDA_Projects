```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Upload the file first if not done
df = pd.read_csv('airline_passenger_satisfaction.csv')
df.head()
```

```python
## Data shape and types
print(df.shape)
print(df.info())

# Null values
print(df.isnull().mean() * 100)
```

```python
df['Arrival Delay']=df['Arrival Delay'].fillna(df['Arrival Delay'].median())
df.isnull().mean() * 100
```

```python
df.describe()
```

```python
categorical_cols = ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'Satisfaction']

for col in categorical_cols:
    print(f"\n{col} value counts:\n")
    print(df[col].value_counts())
```

```python
#
sns.countplot(x='Satisfaction', data=df)
plt.title('Passenger Satisfaction Count')
plt.show()
```

```python
sns.countplot(x='Class',data=df)
plt.title('Distribution by travel class')
plt.show()
```

```python
sns.countplot(x='Customer Type', data=df)
plt.title('Distribution by Customer Type')
plt.show()
```

```python
#Satisfaction by Class:
sns.countplot(x='Class',hue='Satisfaction', data=df)
plt.title('Satisfaction by Class')
plt.show()
```

```python
#Satisfaction by Customer Type:
sns.countplot(x='Customer Type',hue='Satisfaction', data=df)
plt.title('Satisfaction by Customer Type')
plt.show()
```

```python
sns.countplot(x='Type of Travel', hue='Satisfaction', data=df)
plt.title('Satisfaction by Type of Travel')
plt.show()
```

```python
df['Satisfaction_Numeric'] = df['Satisfaction'].map({'satisfied': 1, 'neutral or dissatisfied': 0})
```

```python
import numpy as np
numeric_features=df.select_dtypes(include=[np.number])
numeric_features.columns

corr_matrix=numeric_features.corr()
corr_matrix

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Flight Distance vs Satisfaction:
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Flight Distance', y='Satisfaction_Numeric', data=df)
plt.title('Flight Distance vs Satisfaction')
plt.show()
```

```python
#Departure Delay vs Satisfaction:
plt.figure(figsize=(10,6))
sns.boxplot(x='Satisfaction', y='Departure Delay', data=df)
plt.title('Departure Delay by Satisfaction')
plt.show()
```

```python
#Online Boarding vs Satisfaction:plt.figure(figsize=(10,6))
sns.violinplot(x='Satisfaction', y='Online Boarding', data=df)
plt.title('Online Boarding Rating by Satisfaction')
plt.show()

```

```python
#Pairplots for numerical features:
sns.pairplot(df, hue='Satisfaction', vars=['Flight Distance', 'Departure Delay', 'Arrival Delay', 'Online Boarding', 'Seat Comfort'])
plt.show()
```

```python
#Average satisfaction by rating categories:
rating_features = ['Departure and Arrival Time Convenience', 'Ease of Online Booking', 'Check-in Service', 'Online Boarding',
                   'Gate Location', 'On-board Service', 'Seat Comfort', 'Leg Room Service', 'Cleanliness', 'Food and Drink',
                   'In-flight Service', 'In-flight Wifi Service', 'In-flight Entertainment', 'Baggage Handling']

for feature in rating_features:
    plt.figure(figsize=(8,4))
    sns.barplot(x=feature, y='Satisfaction_Numeric', data=df)
    plt.title(f'Average Satisfaction by {feature}')
    plt.show()
```
