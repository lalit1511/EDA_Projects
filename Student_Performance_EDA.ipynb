```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('/content/StudentsPerformance.csv')

# Show the first few rows
df.head()
```

```python

# Check basic info about the dataset
df.info()

df.isnull().sum()
```

```python
## Summary statistics for numeric columns
df.describe()
```

```python
#Check for missing and duplicated values
# Check for missing values
print("Missing Values:\n", df.isnull().sum())
# Check for duplicated values
print("Missing Values:\n", df.duplicated().sum())
```

```python
# Visualize score distributions
sns.set(style="whitegrid")
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
sns.histplot(df['math score'], kde=True, color='skyblue')
plt.title('Math Score Distribution')

plt.subplot(1,3,2)
sns.histplot(df['reading score'], kde=True, color='salmon')
plt.title('Reading Score Distribution')

plt.subplot(1,3,3)
sns.histplot(df['writing score'], kde=True, color='lightcoral')
plt.title('Writing Score Distribution')

plt.tight_layout()
plt.show()
```

```python
#Compare Scores by Gender
plt.figure(figsize=(15,5))

# Math Score vs Gender
plt.subplot(1,3,1)
sns.boxplot(x='gender',y='math score',data=df,palette='Blues')
plt.title('Math Score by Gender')

# Reading Score vs Gender
plt.subplot(1,3,2)
sns.boxplot(x='gender',y='reading score',data=df,palette='Greens')
plt.title('Reading Score by Gender')

# Writing Score vs Gender
plt.subplot(1,3,3)
sns.boxplot(x='gender',y='writing score',data=df,palette='Reds')
plt.title('Writing Score by Gender')

plt.tight_layout()
plt.show()
```

```python
# Step 7: Analyze Scores by Parental Level of Education
plt.figure(figsize=(16,5))

#math score
plt.subplot(1,3,1)
sns.boxplot(x='parental level of education',y='math score',data=df)
plt.title('Math Score vs Parental Education')
plt.xticks(rotation=45)

#reading score
plt.subplot(1,3,2)
sns.boxplot(x='parental level of education',y='reading score',data=df)
plt.title('Reading Score vs Parental Education')
plt.xticks(rotation=45)

#writing score
plt.subplot(1,3,3)
sns.boxplot(x='parental level of education',y='writing score',data=df)
plt.title('Writing Score vs Parental Education')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Step 8: Impact of Test Preparation Course on Scores
plt.figure(figsize=(16, 5))

# Math Score
plt.subplot(1, 3, 1)
sns.boxplot(x='test preparation course', y='math score', data=df, palette='coolwarm')
plt.title('Math Score vs Test Preparation')

# Reading Score
plt.subplot(1, 3, 2)
sns.boxplot(x='test preparation course', y='reading score', data=df, palette='coolwarm')
plt.title('Reading Score vs Test Preparation')

# Writing Score
plt.subplot(1, 3, 3)
sns.boxplot(x='test preparation course', y='writing score', data=df, palette='coolwarm')
plt.title('Writing Score vs Test Preparation')

plt.tight_layout()
plt.show()
```
