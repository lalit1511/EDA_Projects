```python
import pandas as pd

# Load the files into separate DataFrames
df_2015 = pd.read_csv('2015.csv')
df_2016 = pd.read_csv('2016.csv')
df_2017 = pd.read_csv('2017.csv')
df_2018 = pd.read_csv('2018.csv')
df_2019 = pd.read_csv('2019.csv')

# Check the first few rows of one of the dataframes
df_2015.head()


```

```python
# Check the first few rows of each DataFrame
print("2015 Data:")
print(df_2015.head())

print("2016 Data:")
print(df_2016.head())

# Repeat for other years as needed
```

```python
# Check for missing values and data types
print(df_2015.info())
print(df_2015.isnull().sum())
```

```python
## Concatenate the dataframes into one large dataframe
df_all_years=pd.concat([df_2015,df_2016,df_2017,df_2018,df_2019],ignore_index=True)

df_all_years.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of Happiness Scores across all years
sns.histplot(df_all_years['Score'], kde=True, color='blue')
plt.title('Distribution of Happiness Scores (2015-2019)')
plt.show()

# Correlation heatmap for 2015 data
corr_2015 = df_2015.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_2015, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap for 2015 Data')
plt.show()
```

```python
 #Step 1: Understand the Data Columns
print(df_2015.columns)
print(df_2019.columns)
```

```python
# Standardize column names for 2015
df_2015.rename(columns={
    'Happiness Score': 'Score',
    'Economy (GDP per Capita)': 'GDP per capita',
    'Health (Life Expectancy)': 'Healthy life expectancy',
    'Trust (Government Corruption)': 'Perceptions of corruption',
    'Family': 'Social support',
}, inplace=True)

# Check again
print(df_2015.columns)
```

```python
#📌 Step 3: Basic Analysis
# Top 10 happiest countries in 2015
df_2015[['Country', 'Score']].sort_values(by='Score', ascending=False).head(10)
```

```python
sns.scatterplot(data=df_2015, x='GDP per capita', y='Score')
plt.title('GDP per Capita vs Happiness Score (2015)')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# 3. Basic overview
print("Shape of the dataset:", df_all_years.shape)
print("\nDataset Info:")
print(df_all_years.info())

# 4. Check missing values
print("\nMissing Values:")
print(df_all_years.isnull().sum())

# 5. Check unique years
# Add a 'Year' column to each DataFrame before concatenation
df_2015['Year'] = 2015
df_2016['Year'] = 2016
df_2017['Year'] = 2017
df_2018['Year'] = 2018
df_2019['Year'] = 2019

# Re-create df_all_years with the 'Year' column
df_all_years = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019], ignore_index=True)

print("\nUnique years in data:", df_all_years['Year'].unique())

# 6. Basic statistics
print("\nSummary Statistics:")
print(df_all_years.describe())
```
